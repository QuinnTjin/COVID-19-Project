#### New case weekly statistic

```python
case=case_raw.set_index(['countyFIPS','County Name','State','stateFIPS'])
```

```python
case.head()
```

```python
usacase=case.sum()
```

```python
## in this cell, I am going to compute the weekly case. I find the data every 7 days, and then use the data 7 days later to subtract current data result in the new case number for next week
weeklycasedata={}
numberofweek=1
#new case of the first week 
weeklycasedata[0]=[case.columns[3],usacase[3]]
lastDayOfWeek=3
#index of the last day of the first week 2020-1-25 is index of 3, the Sat of next week starts from index 10
for Sat in range(10,usacase.size,7):
    weeklycasedata[numberofweek]=[case.columns[Sat],usacase[Sat]-usacase[lastDayOfWeek]]
    numberofweek+=1
    lastDayOfWeek=Sat
```

```python
weeklycasedata
```

```python
usweeklycase=pd.DataFrame.from_dict(weeklycasedata,orient='index',columns=['firstweekday','case'])
usweeklycase.head()
```

```python
fig=plt.figure(figsize=(20,6))
plt.bar(range(len(weeklycasedata)),usweeklycase.case,align='center')
plt.xticks(range          (len(weeklycasedata)),list(usweeklycase.firstweekday),rotation=270)
plt.show
```

```python
usweeklycasemean=round(int(usweeklycase.case.mean()))
```

```python
print ("the mean for weekly new case is "+ str(usweeklycasemean) )
```

```python
usweeklycase.case.median()
print ("the median for weekly new case is "+str(round(int(usweeklycase.case.median()))))
```

```python
#we can see mode is 2
usweeklycase.case.mode()
```

### - Identify peak week of the cases and deaths in US and other countries. 

# US CASES PEAK WEEK

```python
US_cases = pd.read_csv('../Data/Input/covid_confirmed_usafacts.csv',na_values=['X'])
US_cases = US_cases[US_cases.countyFIPS>0]
US_cases = US_cases.set_index(['countyFIPS', 'County Name', 'State', 'stateFIPS'])

US_cases_sum = US_cases.sum()

US_cases
```

```python
US_cases_sum
```

```python
# Summarizing weekly cases for US
weekly_cases = {}
week_number = 1

weekly_cases[0] = [US_cases.columns[3], US_cases_sum[3]]
lastDayOfWeek = 3

# Looping through and adding weekly cases for the US
for start in range(10,US_cases_sum.size,7):
    weekly_cases[week_number] = [US_cases.columns[start], US_cases_sum[start] - US_cases_sum[lastDayOfWeek]]
    week_number += 1
    lastDayOfWeek = start
    
US_cases_frame = pd.DataFrame.from_dict(weekly_cases,orient='index',columns=['Week Of','US Cases'])
US_cases_frame = US_cases_frame.set_index(['Week Of'])
US_cases_frame.sort_values(by = 'US Cases', ascending = False).head(1)
```

<b>From the data frame above, we can see that the week of June 25, 2020 was the peak week for US cases which saw 457,339 cases</b>
