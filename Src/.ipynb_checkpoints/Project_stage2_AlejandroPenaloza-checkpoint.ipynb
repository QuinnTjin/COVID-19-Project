<b>Starting with my specific state, NC</b>

```python
# Setting up confirmed cases across North Carolina
confirmed_cases_NC = confirmed_import[confirmed_import['State']=='NC']

# Getting rid of 'Statewide Unallocated' and unneeded stateFIPS column
confirmed_cases_NC = confirmed_cases_NC[confirmed_cases_NC.countyFIPS > 0]
confirmed_cases_NC = confirmed_cases_NC.drop(columns=['stateFIPS'])

# Setting the index
confirmed_cases_NC = confirmed_cases_NC.set_index(['countyFIPS','County Name','State'])
cc_NC_sum = confirmed_cases_NC.sum()
```

```python
cc_NC_sum
```

<b>Next I will do the other 2 states I will be comparing with (South Carolina and Georgia)</b>

```python
# Setting up confirmed cases across South Carolina
confirmed_cases_SC = confirmed_import[confirmed_import['State']=='SC']

# Getting rid of 'Statewide Unallocated' and unneeded stateFIPS column
confirmed_cases_SC = confirmed_cases_SC[confirmed_cases_SC.countyFIPS > 0]
confirmed_cases_SC = confirmed_cases_SC.drop(columns=['stateFIPS'])

# Setting the index
confirmed_cases_SC = confirmed_cases_SC.set_index(['countyFIPS','County Name','State'])
cc_SC_sum = confirmed_cases_SC.sum()
```

```python
confirmed_cases_SC.tail()
```

```python
cc_SC_sum
```

```python
# Setting up confirmed cases across Georgia
confirmed_cases_GA = confirmed_import[confirmed_import['State']=='GA']

# Getting rid of 'Statewide Unallocated' and unneeded stateFIPS column
confirmed_cases_GA = confirmed_cases_GA[confirmed_cases_GA.countyFIPS > 0]
confirmed_cases_GA = confirmed_cases_GA.drop(columns=['stateFIPS'])

# Setting the index
confirmed_cases_GA = confirmed_cases_GA.set_index(['countyFIPS','County Name','State'])
cc_GA_sum = confirmed_cases_GA.sum()
```

```python
confirmed_cases_GA.tail()
```

```python
cc_GA_sum
```

```python
# Summarizing weekly cases for all 3 states
weekly_cases = {}
week_number = 1

weekly_cases[0] = [confirmed_cases_NC.columns[3], cc_NC_sum[3], cc_SC_sum[3], cc_GA_sum[3]]
lastDayOfWeek = 3

# Looping through and adding weekly cases for each state
for start in range(10,cc_NC_sum.size,7):
    weekly_cases[week_number] = [confirmed_cases_NC.columns[start], cc_NC_sum[start] - cc_NC_sum[lastDayOfWeek],
                                cc_SC_sum[start] - cc_SC_sum[lastDayOfWeek], cc_GA_sum[start] - cc_GA_sum[lastDayOfWeek]]
    week_number += 1
    lastDayOfWeek = start
    
cases_frame = pd.DataFrame.from_dict(weekly_cases,orient='index',columns=['Week Of','NC Cases', 'SC Cases', 'GA Cases'])
cases_frame = cases_frame.set_index(['Week Of'])
cases_frame
```

```python
cases_mean = []
cases_median = []
cases_mode = []

# Calculating the mean, median, and mode for cases and putting them into a data frame
for states in cases_frame.columns:
    cases_mean.append(cases_frame[states].mean())
    cases_median.append(cases_frame[states].median())
    cases_mode.append(cases_frame[states].mode())

cases_MMM_frame = pd.DataFrame.from_dict({'State':cases_frame.columns,'Mean':cases_mean,'Median':cases_median, 'Mode':cases_mode})
cases_MMM_frame = cases_MMM_frame.set_index(['State'])
```

<b>Now I will do the same for deaths!</b>

```python
# Setting up deaths across North Carolina
deaths_NC = deaths_import[deaths_import['State']=='NC']

# Getting rid of 'Statewide Unallocated' and unneeded stateFIPS column
deaths_NC = deaths_NC[deaths_NC.countyFIPS > 0]
deaths_NC = deaths_NC.drop(columns=['stateFIPS'])

# Setting the index
deaths_NC = deaths_NC.set_index(['countyFIPS','County Name','State'])
deaths_NC_sum = deaths_NC.sum()

# Setting up deaths across South Carolina
deaths_SC = deaths_import[deaths_import['State']=='SC']

# Getting rid of 'Statewide Unallocated' and unneeded stateFIPS column
deaths_SC = deaths_SC[deaths_SC.countyFIPS > 0]
deaths_SC = deaths_SC.drop(columns=['stateFIPS'])

# Setting the index
deaths_SC = deaths_SC.set_index(['countyFIPS','County Name','State'])
deaths_SC_sum = deaths_SC.sum()

# Setting up deaths across Georgia
deaths_GA = deaths_import[deaths_import['State']=='GA']

# Getting rid of 'Statewide Unallocated' and unneeded stateFIPS column
deaths_GA = deaths_GA[deaths_GA.countyFIPS > 0]
deaths_GA = deaths_GA.drop(columns=['stateFIPS'])

# Setting the index
deaths_GA = deaths_GA.set_index(['countyFIPS','County Name','State'])
deaths_GA_sum = deaths_GA.sum()
```

```python
# Summarizing weekly deaths for all 3 states
weekly_deaths = {}
week_number = 1

weekly_deaths[0] = [deaths_NC.columns[3], deaths_NC_sum[3], deaths_SC_sum[3], deaths_GA_sum[3]]
lastDayOfWeek = 3

# Looping through and adding weekly cases for each state
for start in range(10,deaths_NC_sum.size,7):
    weekly_deaths[week_number] = [deaths_NC.columns[start], deaths_NC_sum[start] - deaths_NC_sum[lastDayOfWeek],
                                deaths_SC_sum[start] - deaths_SC_sum[lastDayOfWeek], deaths_GA_sum[start] - deaths_GA_sum[lastDayOfWeek]]
    week_number += 1
    lastDayOfWeek = start
    
deaths_frame = pd.DataFrame.from_dict(weekly_deaths,orient='index',columns=['Week Of','NC Deaths', 'SC Deaths', 'GA Deaths'])
deaths_frame = deaths_frame.set_index(['Week Of'])
deaths_frame
```

```python
deaths_mean = []
deaths_median = []
deaths_mode = []

# Calculating the mean, median, and mode of deaths and putting them into a data frame
for states in deaths_frame.columns:
    deaths_mean.append(deaths_frame[states].mean())
    deaths_median.append(deaths_frame[states].median())
    deaths_mode.append(deaths_frame[states].mode())

deaths_MMM_frame = pd.DataFrame.from_dict({'State':deaths_frame.columns,'Mean':deaths_mean,'Median':deaths_median, 'Mode':deaths_mode})
deaths_MMM_frame = deaths_MMM_frame.set_index(['State'])
deaths_MMM_frame
```
