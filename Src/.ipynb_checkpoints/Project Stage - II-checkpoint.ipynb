### 2.2 Model a poission distribution of COVID-19 cases and deaths of a state and compare to other 5 states. Number of new cases and deaths per 1000 populations.

```python
case_raw.head()
```

```python
popu_raw.head()
```

```python
case_popu_raw=popu_raw.merge(case_raw)
death_popu_raw=popu_raw.merge(death_raw)
```

```python
case_popu_raw.head()
```

```python
death_popu_raw.head()
```

#### Get the table of  daily new case and death for TX, then I model a poission distribution of new case death for TX and compare it with another  5 states, NY, NC, SC, CA, AL, normalize by population. Same as part 2.1

```python
def daily_newcase_per_10000ppl(state):
    tem=pd.DataFrame(case_popu_raw[case_popu_raw.State==state].sum()).T    
    statepopulation= tem.iloc[0,3]
    for x in tem.columns[5:]:
        tem[x]=round(tem[x]/statepopulation*10000)          ## normalize by population      
    tem=tem[tem.columns[5:]]
    tem=tem.diff(axis=1)    
    tem=tem.fillna(0)
    tem.index=['daily new case per 10000 population-'+state]    
    tem=tem.T
    return tem
```

```python
def daily_death_per_million(state):
    tem=pd.DataFrame(death_popu_raw[death_popu_raw.State==state].sum()).T    
    statepopulation= tem.iloc[0,3]
    for x in tem.columns[5:]:
        tem[x]=round(tem[x]/statepopulation*1000000)          ## normalize by population
    tem=tem[tem.columns[5:]]
    tem=tem.diff(axis=1)    
    tem=tem.fillna(0)
    tem.index=['daily death per million population-'+state]
    tem=tem.T
    return tem
```

#### Texas daily new case and death, we have done the daily new case above. 

```python
daily_newcase_per_10000ppl_TX=daily_newcase_per_10000ppl('TX')
```

```python
daily_newcase_per_10000ppl_TX.describe()
```

```python
daily_newcase_per_10000ppl_TX.hist(bins=7,density=True)
plt.plot(np.linspace(0,7,num=8),poisson.pmf(np.linspace(0,7,num=8),1),'r')
```

```python
daily_death_per_million_TX=daily_death_per_million('TX')
```

```python
daily_death_per_million_TX.describe()
```

```python
daily_death_per_million_TX.hist(bins=20,density=True)
plt.plot(np.linspace(0,83,num=84),poisson.pmf(np.linspace(0,83,num=84),2),'r')
```

#### We can see the lambda value is 1 for both newcase_per_10000 people and lambda is around 2 for  the death per billion people of TX. Then we those two possion distribution to compare with other 5 states.

```python
for x in ['NY', 'NC', 'SC', 'CA', 'AL']:
    tablename=daily_newcase_per_10000ppl(x)
    bins=int(tablename.max())
    tablename.hist(bins=bins,density=True,ec='black')
    plt.plot(np.linspace(0,tablename.max(),num=tablename.max()+1),poisson.pmf(np.linspace(0,tablename.max(),num=tablename.max()+1),1),'r')
```

```python
for x in ['NY', 'NC', 'SC', 'CA', 'AL']:
    tablename=daily_death_per_million(x)
    bins=int(tablename.max())
    tablename.hist(bins=bins,density=True,ec='black')
    plt.plot(np.linspace(0,tablename.max(),num=tablename.max()+1),poisson.pmf(np.linspace(0,tablename.max(),num=tablename.max()+1),2),'r')
```

- ##  Model poission distributions for North Carolina counties COVID-19 in cases and deaths. Number of new cases and deaths per 1000 populations.

```python
case_popu_raw.head()
```

```python
death_popu_raw.head()
```

```python
case_popu_raw_NC=case_popu_raw[case_popu_raw.State=='NC']
death_popu_raw_NC=death_popu_raw[death_popu_raw.State=='NC']
```

```python
top5_case_NC_counties=case_popu_raw_NC.sort_values(by=case_popu_raw_NC.columns[-1],ascending=False).iloc[:5,:]
top5_case_NC_counties
```

```python
# find the counties of top 5 cases 
countylist_case=[x for x in top5_case_NC_counties['County Name']]
countylist_case
```

```python
top5_death_NC_counties=case_popu_raw_NC.sort_values(by=case_popu_raw_NC.columns[-1],ascending=False).iloc[:5,:]
top5_death_NC_counties
```

```python
# find the counties of top 5 death
countylist_death=[x for x in top5_death_NC_counties['County Name']]
countylist_death
```

```python
def daily_newcase_per_100000_NC_county(county):    
    tem=top5_case_NC_counties[top5_case_NC_counties['County Name']==county]
    countypopulation= tem.iloc[0,3]
    for x in tem.columns[5:]:
        tem[x]=int(tem[x]/countypopulation*100000)          ## normalize by population     
    tem=tem[tem.columns[5:]]
    tem=tem.diff(axis=1)    
    tem=tem.fillna(0)
    tem.index=['daily new case per 100000 population-'+county]    
    tem=tem.T
    return tem
```

```python
def daily_death_per_million_NC_county(county):
    tem=top5_death_NC_counties[top5_death_NC_counties['County Name']==county]    
    countypopulation= tem.iloc[0,3]
    for x in tem.columns[5:]:
        tem[x]=int(tem[x]/countypopulation*100000)          ## normalize by population    
    tem=tem[tem.columns[5:]]
    tem=tem.diff(axis=1)    
    tem=tem.fillna(0)
    tem.index=['daily death per million population-'+county]
    tem=tem.T
    return tem
```

```python
for x in countylist_case:
    tablename=daily_newcase_per_100000_NC_county(x)
    mean=round(tablename.mean())
    tablename.hist(bins=20,density=True,ec='black')
    plt.plot(np.linspace(0,tablename.max(),num=tablename.max()+1),poisson.pmf(np.linspace(0,tablename.max(),num=tablename.max()+1),mean),'r')
```

```python
for x in countylist_death:
    tablename=daily_death_per_million_NC_county(x)
    mean=round(tablename.mean())
    tablename.hist(bins=20,density=True,ec='black')
    plt.plot(np.linspace(0,tablename.max(),num=tablename.max()+1),poisson.pmf(np.linspace(0,tablename.max(),num=tablename.max()+1),mean),'r')
```
