# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read csv files

```python
# COVID-19 dataset
conf_add = "../Data/covid_confirmed_usafacts.csv"
deaths_add = "../Data/covid_deaths_usafacts.csv"
popu_add = "../Data/covid_county_population_usafacts.csv"
confirmed = pd.read_csv(conf_add)
deaths = pd.read_csv(deaths_add)
population = pd.read_csv(popu_add)

# Enrichment dataset
beds_add = "../Data/Definitive_Healthcare_Hospital_Beds_0914.csv"
hospital_beds = pd.read_csv(beds_add)
```

# Display dataframe

```python
#COVID-19 dataset

confirmed_index = confirmed.set_index(['countyFIPS','State'])

print('Confirmed shape',confirmed_index.shape)

confirmed_index.head()
```

```python
confirmed.countyFIPS.value_counts()
```

```python
deaths_index = deaths.set_index(['countyFIPS','State'])

print('Deaths shape',deaths_index.shape)

deaths_index.head()
```

```python
population_index = population.set_index(['countyFIPS','State'])

print('Population shape',population_index.shape)

population_index.head()
```

# Merging COVID Dataset

```python
# Merging Step 1: merging population df with confirmed df, inner join
merge_1 = population.merge(confirmed)

# Converting the merge_1 dataframe to a long format
merge_11 = pd.melt(merge_1, id_vars=['countyFIPS','County Name','State','stateFIPS','population'], 
        var_name='dates', value_name='confirmed_cases')
merge_11_index = merge_11.set_index(['countyFIPS','State','dates']).sort_values(['countyFIPS','State','dates'])

# Converting deaths dataframe to long format
deaths_1 = pd.melt(deaths, id_vars=['countyFIPS','County Name','State','stateFIPS'], 
        var_name='dates', value_name='deaths')
deaths_1_index = deaths_1.set_index(['countyFIPS','State','dates']).sort_values(['countyFIPS','State','dates'])

# Merging Step 2: merging the resultant of step 1 with the deaths df, inner join
merge_final = merge_11.merge(deaths_1)
covid_merge_final = merge_final.sort_values(['countyFIPS','State','dates'])

print('Final merge file shape: ',covid_merge_final.shape)
covid_merge_final.head(10)
```

# Trends for confirmed cases last week 

```python
confirmed_VA = confirmed[confirmed.State=='VA']
print('Confirmed shape in VA:',confirmed_VA.shape)
confirmed_VA.head()
```

```python
confirmed_last_week = confirmed_VA.iloc[:,[0,1,2,3,-1,-2,-3,-4,-5,-6,-7]]

# convert confirmed table to long format
confirmed_last_week = pd.melt(confirmed_last_week, id_vars=['countyFIPS','County Name','State','stateFIPS'], 
        var_name='dates', value_name='confirmed')
confirmed_last_week['dates'] = pd.to_datetime(deaths_last_week['dates'])
confirmed_last_week.sort_values(['dates','countyFIPS'])
print('Last week confirmed cases shape:',confirmed_last_week.shape)
confirmed_last_week.head()
```

```python
# Grouping by dates and summing the confirmed cases across all counties
confirmed_groups=confirmed_last_week.groupby('dates')['confirmed'].sum()
confirmed_groups
confirmed_groups.plot()
plt.title('Total confirmed cases across counties each day - past week')
```

```python
# Group by countyFIPS and summing the confirmed cases for each county
confirmed_groups=confirmed_last_week[confirmed_last_week.countyFIPS>0].groupby('countyFIPS')['confirmed'].sum()
ax=plt.subplot(111)
confirmed_groups.plot.bar(ax=ax)
ax.set_ylabel('confirmed')
```

## *Confirmed cases have been increasing continuously for the past one week which can be seen in the first plot. From the second plot we can see that few counties have much higher number of confirmed cases than others in the past week.

# Trends of deaths for the past one week

```python
deaths_VA = deaths[deaths.State=='VA']
print('Deaths shape in VA:',deaths_VA.shape)
deaths_VA.head()
```

```python
deaths_last_week = deaths_VA.iloc[:,[0,1,2,3,-1,-2,-3,-4,-5,-6,-7]]

# convert deaths table to long format
deaths_last_week = pd.melt(deaths_last_week, id_vars=['countyFIPS','County Name','State','stateFIPS'], 
        var_name='dates', value_name='deaths')
deaths_last_week['dates'] = pd.to_datetime(deaths_last_week['dates'])
deaths_last_week.sort_values(['dates','countyFIPS'])
print('Last week deaths shape:',deaths_last_week.shape)
deaths_last_week.head()
```

```python
# Grouping by dates and summing the deaths for across all counties
deaths_groups=deaths_last_week.groupby('dates')['deaths'].sum()
deaths_groups
deaths_groups.plot()
plt.title('Total deaths across counties each day - past week')
```

```python
deaths_groups=deaths_last_week[deaths_last_week.countyFIPS>0].groupby('countyFIPS')['deaths'].sum()
# deaths_groups.plot()
ax=plt.subplot(111)
deaths_groups.plot.bar(ax=ax)
ax.set_ylabel('deaths')
```

# Dealing with missing values in enrichment dataset

```python
print('Hospital Beds dataset shape',hospital_beds.shape)

# Finding out if there are any missing/null vlaues in FIPS, CNTY_FIPS,STATE_FIPS
print('Are there any missing/null values?',hospital_beds['FIPS'].isnull().values.any())

# How many missing values
missing_cnt = hospital_beds['FIPS'].isnull().sum()
print('Missing count FIPS, CNTY_FIPS,STATE_FIPS:',missing_cnt)

# shape of missing values
missing_rows = hospital_beds[hospital_beds['FIPS'].isnull()]
print('\n\n',missing_rows.head(1))

# shape of records without any missing county FIPS information
records_not_missing = hospital_beds[hospital_beds['FIPS'].notna()]

print('\nRecords with missing county FIPS code:',missing_rows.shape[0])
print('Records with no missing county FIPS code:',records_not_missing.shape[0])

```

# converting the datatype of FIPS to integer from float

```python
records_not_missing['FIPS'] = records_not_missing['FIPS'].astype(np.int64)

print('Shape:',records_not_missing.shape)
records_not_missing.sort_values(['FIPS'])
```

# Merging COVID dataset with hospital beds dataset using "countyFIPS" in COVID dataset and "FIPS" in enrichment dataset

```python
covid_beds = covid_merge_final.merge(records_not_missing,left_on=['countyFIPS'], right_on=['FIPS'])
print('COVID merged with Hospital beds shape:',covid_beds.shape)
covid_beds.sort_values(['countyFIPS','dates']).head()
```
